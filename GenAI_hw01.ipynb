```python
%matplotlib inline


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#將每個組成"貓"的函數列出
def f1(x):
    return 0*x

def f2(x):
    return -0.7 * (x - 9.9)**2 + 10

def f3(x):
    return 0.5 * np.sqrt(x)

def f4(x):
    return np.log(200) / np.log(x - 5) + 1

def f5(x):
    return (x - 6.7)**3 + x - 2

def f6(x):
    return -(0.5 * x - 3.1)**2 + 10

def f7(x):
    return -2.3 * np.sqrt(x - 6) + 11

def f8(x):
    return 0.5 * x + 8

def f9(x):
    return -0.4 * (x - 9.1)**2 + 10.5

#def f10(x):   x=7

def f11(x):
    return -(4**(x - 9.6)) + 12.4

def f12(x):
    return -2 / (2 + 10**(4*x - 27)) + 12.2

def f13(x):
    return np.abs(2*x - 15.5)*(-1) + 13.5

def f14(x):
    return np.abs(3*x - 27)*(-1) + 13.5

def outputX(y):
    return ((y-1)**2)*(-0.12)+13.4
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))  #plot size

x = np.linspace(0, 13.4, 500) #array
plt.plot(x, f1(x))

x = np.linspace(6.37, 13.3, 500)
plt.plot(x, f2(x))

x = np.linspace(0, 10, 500)
plt.plot(x, f3(x))

x = np.linspace(6.995, 7.65, 500)
plt.plot(x, f4(x))

x = np.linspace(5.4, 7.65, 500)
plt.plot(x, f5(x))

x = np.linspace(6.2, 7, 500)
plt.plot(x, f6(x))

x = np.linspace(6, 6.19, 500)
plt.plot(x, f7(x))

x = np.linspace(6, 6.5, 500)
plt.plot(x, f8(x))

x = np.linspace(6.98, 10.4, 500)
plt.plot(x, f9(x))

x = np.linspace(7, 7, 500)
y = np.linspace(8.7, 9.8, 500)
plt.plot(x, y)

x = np.linspace(8.39, 10.24, 500)
plt.plot(x, f11(x))

x = np.linspace(6.48, 7, 500)
plt.plot(x, f12(x))

x = np.linspace(7, 8.5, 500)
plt.plot(x, f13(x))

x = np.linspace(8.56, 9.8, 500)
plt.plot(x, f14(x))

y = np.linspace(-5, 7, 500)    #間距 = 7-(-5)/499
x = outputX(y)

filter = x >= 13.29     #output a Boolean array  [20, 10, 30, 35, 5]->[T, F, T, T, F]
# print(filter)
xf = x[filter]          #mask filter on X and Y
yf = y[filter]

plt.plot(xf, yf);

plt.title('Plot of Cat')
plt.axis('off')
# plt.grid(True)
# plt.legend()   # definition of legend = caption , to show label from plt.plot (functions)

```
